```python
df.drop_duplicates(inplace=True)
```

```python
df.shape
```

```python
df['premiere']=pd.to_datetime(df['premiere'])
```

```python
df['premiere_month']=df['premiere'].dt.month
```

```python
df['premiere_day']=df['premiere'].dt.day
```

```python
df['premiere_year']=df['premiere'].dt.year
```

```python
df.hist(bins=40,grid=True,color='chocolate',figsize=(15,9))
plt.show()
```

```python
df.head(8)
```

```python
df.language.value_counts().head(10).plot(kind='pie')
plt.title('Types of Language in Movies')
plt.tight_layout()

```

```python
plt.figure(figsize=(10,5))
df.genre.value_counts().head(15).plot(kind='bar',color=sns.color_palette('husl'))
plt.title('Types of Genre of Movies')
plt.xticks(rotation=45)
plt.tight_layout()
```

```python
yearly_count=df['premiere_year'].value_counts().sort_index(ascending=False)
plt.figure(figsize=(10,6))
sns.barplot(x=yearly_count.index,y=yearly_count.values, hue=yearly_count.index, palette='mako',legend=False)
plt.title('Yearly Count Premier')
plt.tight_layout()
plt.xlabel('Premiere Year')
plt.ylabel('Count')

```

```python
plt.figure(figsize=(12,4))
sns.barplot(data=df,x='year',y='runtime',palette='magma',legend=False,hue='year')
plt.title('Yearly Run Time')
plt.tight_layout()
```

```python
plt.figure(figsize=(9,4))
sns.lineplot(data=df,x='imdb_score',y='runtime',color='olive')
plt.title('Score on the Basis of Runtime')
plt.tight_layout()
```

```python
plt.figure(figsize=(9,4))
top_6_movies=df.sort_values(by='imdb_score',ascending=False).head(6)
sns.barplot(x=top_6_movies['title'],y=top_6_movies['imdb_score'],palette='icefire',hue=top_6_movies['title'],legend=False)
plt.xlabel('Movie Name')
plt.ylabel('Score')
plt.xticks(rotation=35)
plt.tight_layout()
```

```python
plt.figure(figsize=(17,5))
sns.barplot(x='language',y='imdb_score',data=df,palette='rocket',hue='language',legend=False)
plt.title('Score on the Basis of Language')
plt.xlabel('Language')
plt.ylabel('Score')
plt.xticks(rotation=90)
plt.tight_layout()
```
